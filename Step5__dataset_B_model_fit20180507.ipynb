```python
# 列表保留小数点3位子函数
def func_list_keep_num_demical(list_long_demical,num_demical):
    list_num_demical = []
    for i in range(len(list_long_demical)):
        list_num_demical.append(round(list_long_demical[i],num_demical)) 
    return list_num_demical
```

```python
list_y_pred_1_three_demical = func_list_keep_num_demical(list_y_pred_1,3)
list_y_pred_2_three_demical = func_list_keep_num_demical(list_y_pred_2,3)
list_y_pred_3_three_demical = func_list_keep_num_demical(list_y_pred_3,3)
list_y_pred_4_three_demical = func_list_keep_num_demical(list_y_pred_4,3)
list_y_pred_5_three_demical = func_list_keep_num_demical(list_y_pred_5,3)
```

```python
# list 填入 dataframe 子函数
import pandas as pd
def func_add_list_to_df_columns(df_under_add,list_add,str_column):
    se = pd.Series(list_add)
    df_under_add[str_column] = se.values
    return df_under_add
```
