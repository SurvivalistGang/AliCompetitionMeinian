## 从全体特征训练集中根据官方测试集筛选出对应的测试集
    总体思路
    1 仿造a榜数据特征值，从b榜从筛选出相同的107个特征值
    2 转float
    3 根据 b榜 vid顺序重新排序

```python
import pandas as pd
import numpy as np
```

```python
dataset_test_label_read = pd.read_csv('meinian_round1_test_b_20180505.csv', encoding="gb2312")
dataset_test_label = dataset_test_label_read 
print("\n", "dataset_test_label_read.shape: ","\n", dataset_test_label_read.shape) 
print("\n", "dataset_test_label_read.head(5): ","\n", dataset_test_label_read.head(5)) 
```

```python
# 功能：从全体样本中筛选出部分样本
# str_feature_type：用于筛选的特征值
def select_sample_has_label(df_all_sample,df_part_sample,str_feature_type):
    # 获取标签训练集的 ID list  即 vid 的 list
    list_vid_train_label =df_part_sample[str_feature_type].tolist()
    # 根据ID号，筛选出38199个样本，再组成一个新的 dataframe
    df_selected_via_vid = df_all_sample.loc[df_all_sample[str_feature_type].isin(list_vid_train_label)]
    # 之前的 index 还是断断续续的，扔掉之前的，重新从0开始排序
    df_selected_via_vid_index_resetted = df_selected_via_vid.reset_index(drop=True)
    return df_selected_via_vid_index_resetted
```

```python
# 读取完整训练特征集
dataset_train_feature = pd.read_csv("tmp.csv")
```

```python
dataset_test_feature = select_sample_has_label(dataset_train_feature,
                                               dataset_test_label,
                                              'vid')
print("\n", "dataset_test_feature.shape: ","\n", dataset_test_feature.shape) 
print("\n", "dataset_test_feature: ","\n", dataset_test_feature) 
```

```python
# 子函数，用于删除从 csv 中读入文件的第一行
def func_delete_Unnamed0(df_from_csv):
    df_delete_Unnamed0 =df_from_csv.drop(['Unnamed: 0'],axis=1)
    return df_delete_Unnamed0
```

```python
# 导入 预处理好的特征训练集

dataset_A_test_feature_read = pd.read_csv('dataset_train_feature_three_20180507.csv')
dataset_A_test_feature = dataset_A_test_feature_read
print("\n", "dataset_A_test_feature.shape: ","\n", dataset_A_test_feature.shape) 
```

```python
# 测试集预处理
# 特征值筛选，2796 个  减少到 136个
def func_df_select_feature(df_under,df_model):
    list_feature_model =df_model.columns.tolist()
    df_result = df_under[list_feature_model]
    return df_result

dataset_test_feature_select_columns = func_df_select_feature(dataset_test_feature,
                                                             dataset_A_test_feature )

print("\n", "dataset_test_feature_select_columns.shape: ","\n", dataset_test_feature_select_columns.shape) 
print("\n", "dataset_test_feature_select_columns: ","\n",dataset_test_feature_select_columns) 




```

```python
# 转float
# 转float类型
def change_not_float_to_nan(check_str):
    try:
        float(check_str)
        str_c = float(check_str)
        return str_c
    except ValueError:
        check_str = np.nan
        return check_str
    
def dataset_change_not_float_to_nan(df_under):
    df_changed = df_under
    for i in range(df_under.shape[0]):
        # 这里'vid' 特征是 str类型，但是要保存'vid'特征，'vid'特征是第0列，所以要从1开始循环
        for j in range(1,df_under.shape[1]):
            df_changed.iat[i, j] = change_not_float_to_nan(df_under.loc[i].values[j])

    return df_changed


dataset_test_feature_all_float = dataset_change_not_float_to_nan(dataset_test_feature_select_columns)
print("\n", "dataset_test_feature_all_float.shape: ","\n", dataset_test_feature_all_float.shape) 
print("\n", "dataset_test_feature_all_float.head(5): ","\n",dataset_test_feature_all_float.head(5)) 
```

### 根据测试标签集ID顺序对测试特征集进行重新排序

```python
def print_df_info(df):
    print("\n", "\n", df.shape) 
    print("\n","\n",df.head(5))
    
def func_sort_df(df_under_sorted,df_model,str_sort_key):
    list_sorter=  df_model[str_sort_key].tolist()
    dict_sorterIndex = dict(zip(list_sorter,range(len(list_sorter))))
    str_sort_key_rank = str_sort_key + '_rank'
    df_under_sorted[str_sort_key_rank] = df_under_sorted[str_sort_key].map(dict_sorterIndex)
    df_under_sorted.sort_values([str_sort_key_rank], \
        ascending = [True ], inplace = True)
    df_under_sorted.drop(str_sort_key_rank, 1, inplace = True)
    #df_sorted = df_under_sorted[df_under_sorted[str_sort_key].isin(list_model_sort_key)]
    return df_under_sorted


dataset_test_feature_sorted = func_sort_df(dataset_test_feature_all_float,
                                          dataset_test_label,
                                          'vid')
print("\n","dataset_test_feature_sorted" ,"\n" ) 
print_df_info(dataset_test_feature_sorted)

print("\n","dataset_test_label" ,"\n" ) 
print_df_info(dataset_test_label)
```

```python
# 处理好的测试集写入
dataset_test_feature_sorted.to_csv('dataset_B_test_featrue_preprocessed_20180507.csv',index = False)
```
